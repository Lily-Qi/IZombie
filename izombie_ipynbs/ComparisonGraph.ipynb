```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df1 = pd.read_csv('DQN_sun300.csv')
df2 = pd.read_csv('DQN300.csv')

plt.figure(figsize=(10, 5))
plt.plot(df1['Step Count'], df1['Average Remaining Suns'], label='Only Sun', color='tab:blue')
plt.plot(df2['Step Count'], df2['Average Remaining Suns'], label='Sun + Brain', color='tab:red')

# Adding legends, labels, and title
plt.legend()
plt.xlabel('Step Count')
plt.ylabel('Average Remaining Suns')
plt.title('Average Remaining Suns Over Steps for DQN + MLP with Different Reward Models')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df1 = pd.read_csv('RainbowDQN.csv', header=None)
df1.columns = ['Step Count', 'Winning Rate', 'Average Remaining Suns']

df2 = pd.read_csv('DQN.csv')
df3 = pd.read_csv('RainbowDQN+LSTM.csv')
df4 = pd.read_csv('dqn+lstm.csv')

plt.figure(figsize=(5, 5))
plt.plot(df2['Step Count'], df2['Average Remaining Suns'], label='DDQN', color='tab:gray')
plt.plot(df1['Step Count'], df1['Average Remaining Suns'], label='Rainbow DQN', color='tab:red')
plt.plot(df3['Step Count'], df3['Average Remaining Suns'], label='Rainbow DQN + LSTM', color='tab:green')
plt.plot(df4['Step Count'], df4['Average Remaining Suns'], label='DDQN + LSTM', color='tab:blue')
plt.axhline(y=1104.76, color='tan', linestyle='--', label='Human Baseline')

# Adding legends, labels, and title
plt.legend()
plt.xlabel('Step Count')
plt.ylabel('Average Remaining Suns')
plt.title('Average (Winning) Remaining Suns Comparisons')
current_ylim = plt.ylim()  # Get the current y-axis limits
plt.ylim(current_ylim[0], current_ylim[1] * 1.145)

plt.legend(loc='upper right', fontsize=6)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df1 = pd.read_csv('RainbowDQN(n=3).csv')
df2 = pd.read_csv('RainbowDQN(n=10).csv')
df3 = pd.read_csv('RainbowDQN(n=3batch=128).csv')
df1['Winning Rate'] = df1['Winning Rate'] * 100
df2['Winning Rate'] = df2['Winning Rate'] * 100
df3['Winning Rate'] = df3['Winning Rate'] * 100

plt.figure(figsize=(5, 5))
plt.plot(df3['Step Count'], df3['Winning Rate'], label='n=3, batch size=128', color='tab:green')
plt.plot(df1['Step Count'], df1['Winning Rate'], label='n=3, batch size=256', color='tab:gray')
plt.plot(df2['Step Count'], df2['Winning Rate'], label='n=10, batch size=256', color='tab:blue')

# Adding legends, labels, and title
plt.legend()
plt.xlabel('Step Count')
plt.ylabel('Average Remaining Suns')
plt.title('Winning Rates Comparisons Using RainbowDQN')
current_ylim = plt.ylim()  # Get the current y-axis limits
plt.ylim(current_ylim[0], current_ylim[1] * 1.145)

plt.legend(loc='upper right', fontsize=6)
plt.show()
```
